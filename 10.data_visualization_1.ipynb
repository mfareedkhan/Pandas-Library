```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# import titanc data from seabron library
df = sns.load_dataset('titanic')
```

```python
# Plots using Pandas:
# Scatter Plot
df.plot(kind='scatter', x='age', y='fare')
```

```python
# changing color using hex code:
df.plot(kind='scatter', x='age', y='fare', color='#203864')
```

```python
# Plots using Pandas:
# Line Plot
df.plot(kind='line', x='age', y='fare', color='#203864')
```

```python
df.groupby('sex')['age'].mean()
```

```python
df.groupby('sex')['age'].mean().plot(kind='bar', color='#203864')
```

```python
df.groupby('sex')['age'].mean().plot(kind='bar', color='#203864')
plt.ylabel('Average Age')
plt.xlabel('Gender')
```

```python
# Seaborn Plots:
# Scatter Plot
sns.scatterplot(x='age', y='fare', data=df, color='#203864')
```

```python
# Seaborn Plots:
# Line Plot
sns.lineplot(x='age', y='fare', data=df, color='#203864')
```

```python
# Seaborn Plots:
# Bar Plot
sns.barplot(data=df, x='sex', y='fare', color='#f4c708')
```

```python
# Seaborn Plots:
# Bar Plot
sns.barplot(data=df, x='sex', y='fare', hue='class', color='#203864')
```

```python
# Seaborn Plots:
# Box Plot
sns.boxplot(data=df, x='sex', y='age', hue='class', color='#f4c708')
```

```python
sns.scatterplot(data=df, x='fare', y='age', hue='sex', color='#f4c708')
# both axis must be numeric for scatter plot
```

```python
# Seaborn Plots:
# Histogram
sns.distplot(df['age'], bins=10, kde=False, color='#203864')
```

```python
import plotly.graph_objects as go

import pandas as pd

# Read data from a csv
z_data = pd.read_csv('https://raw.githubusercontent.com/plotly/datasets/master/api_docs/mt_bruno_elevation.csv')

fig = go.Figure(data=[go.Surface(z=z_data.values)])

fig.update_layout(title='Mt Bruno Elevation', autosize=False,
                  width=500, height=500,
                  margin=dict(l=65, r=50, b=65, t=90))

fig.show()
```

#### Observations:
- panda can read data from online link as well (github link)
- `pd.read_csv(link)` is used to read data from online link
- `go.Figure()` is used to create a figure
- `go.Surface()` is used to create a surface plot
- `fig.update_layout()` is used to update the layout of the plot
- `fig.show()` is used to show the plot
--- 
- this plot shows the elevation of Mt Bruno

```python
# Animated plot using plotly express:
import plotly.express as px
df = px.data.gapminder()
px.scatter(df, x="gdpPercap", y="lifeExp", animation_frame="year", animation_group="country",
           size="pop", color="continent", hover_name="country",
           log_x=True, size_max=55, range_x=[100,100000], range_y=[25,90])
```

#### Observations:
- gapmider is a dataset in plotly library
- animation_frame is the column that we want to animate
- animation_group is the column that we want to group by
- size is the size of the dots
- color is the color of the dots
- hover_name is the name that we want to show when we hover over the dots
- log_x is the log of x axis (logarithmic scale)
- size_max is the maximum size of the dots
- range_x is the range of x axis
- range_y is the range of y axis
- we can also use 'animation_group' instead of 'color' to group the dots 
--- 
- this data shows the life expectancy and GDP per capita of different countries
- we can see the change in life expectancy and GDP per capita over the years
