### 1. Bar Chart:
#### Frequency of Each Order Status

```python
import matplotlib.pyplot as plt

def create_bar_chart(df, column, title, output_file):
    # Get the frequency of each order status
    status_counts = df[column].value_counts()

    # Create the bar chart
    status_counts.plot(kind='bar')

    # Set the title and labels
    plt.title(title)
    plt.xlabel(column)
    plt.ylabel('Frequency')

    # Save the plot to a file
    plt.savefig(output_file, dpi=300, bbox_inches='tight')

    # Display the plot
    plt.show()

# Use the function to create the bar chart
create_bar_chart(df, 'status', 'Frequency of Each Order Status', './outputs/plots/bar_chart.png')
```

#### Interpretation:

- The bar chart represents the frequency of each order status in the dataset.
- A high frequency of 'complete' orders indicates a high order fulfillment rate.
- A high frequency of 'canceled' or 'order_refunded' statuses could suggest issues with order fulfillment.
- The specific insights will depend on the actual data in your chart.

### 2. Pie Chart:

```python
def create_pie_chart(df, column, title, output_file):
    # Get the frequency of each payment method
    payment_counts = df[column].value_counts()

    # Create the pie chart
    plt.figure(figsize=(10, 8))
    plt.pie(payment_counts, labels=payment_counts.index, autopct='%1.1f%%')

    # Set the title
    plt.title(title)

    # Save the plot to a file
    plt.savefig(output_file, dpi=300, bbox_inches='tight')

    # Display the plot
    plt.show()

# Use the function to create the pie chart
create_pie_chart(df, 'payment_method', 'Distribution of Payment Methods', './outputs/plots/pie_chart.png')
```

#### Interpretation:
- The pie chart represents the distribution of different payment methods used in the dataset.
- Each slice of the pie corresponds to a different payment method. The size of the slice indicates the proportion of transactions made using that method.
- A larger slice for a particular payment method indicates that it is more commonly used by customers.
- The specific insights will depend on the actual data in your chart. For example, if one payment method dominates, it might suggest a preference or convenience for that method among your customers.

### 3. Histogram
#### Distribution of Order Prices:

```python
def create_histogram(df, column, title, output_file):
    # Create the histogram
    plt.figure(figsize=(10, 8))
    plt.hist(df[column].dropna(), bins=30, edgecolor='black')

    # Set the title and labels
    plt.title(title)
    plt.xlabel(column)
    plt.ylabel('Frequency')

    # Save the plot to a file
    plt.savefig(output_file, dpi=300, bbox_inches='tight')

    # Display the plot
    plt.show()

# Use the function to create the histogram
create_histogram(df, 'grand_total', 'Distribution of Order Prices', './outputs/plots/histogram.png')
```

#### Interpretation:
- The histogram represents the distribution of order prices in the dataset.
- The x-axis represents the price range, while the y-axis represents the frequency of orders in each price range.
- The height of each bar corresponds to the number of orders in that price range.
- The specific insights will depend on the actual data in your chart. For example, a skew towards lower prices could indicate that most customers prefer cheaper items, while a more uniform distribution might suggest a wider range of customer preferences.

### 4. Scatter Plot:
#### Relationship b/w Price & Quantity Ordered:

```python
import seaborn as sns

def create_scatter_plot(df, x_column, y_column, title, output_file):
    # Create the scatter plot
    plt.figure(figsize=(10, 8))
    sns.scatterplot(data=df, x=x_column, y=y_column)

    # Set the title and labels
    plt.title(title)
    plt.xlabel(x_column)
    plt.ylabel(y_column)

    # Save the plot to a file
    plt.savefig(output_file, dpi=300, bbox_inches='tight')

    # Display the plot
    plt.show()

# Use the function to create the scatter plot
create_scatter_plot(df, 'price', 'qty_ordered', 'Relationship between Price and Quantity Ordered', './outputs/plots/scatter_plot.png')
```

### 5. Box Plot


```python
def create_box_plot(df, column, title, output_file):
    # Create the box plot
    plt.figure(figsize=(10, 8))
    sns.boxplot(data=df, x=column)

    # Set the title and labels
    plt.title(title)
    plt.xlabel(column)

    # Save the plot to a file
    plt.savefig(output_file, dpi=300, bbox_inches='tight')

    # Display the plot
    plt.show()

# Use the function to create the box plot
create_box_plot(df, 'price', 'Distribution of Prices', './outputs/plots/box_plot.png')
```
